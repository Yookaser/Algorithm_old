```python
# 1021
class deque:
    def __init__(self, X=[]):
        self.len = len(X)
        self.deque = X
        
    def push_front(self, num):
        self.deque.insert(0, num)
        self.len += 1
        
    def push_back(self, num):
        self.deque.append(num)
        self.len += 1
        
    def pop_front(self):
        if self.len == 0: return -1
        self.len -= 1
        return self.deque.pop(0)
    
    def pop_back(self):
        if self.len == 0: return -1
        self.len -= 1
        return self.deque.pop()
    
    def size(self):
        return self.len
    
    def empty(self):
        if self.len == 0: return 1
        else: return 0
        
    def front(self):
        if self.len == 0: return -1
        return self.deque[0]
    
    def back(self):
        if self.len == 0: return -1
        return self.deque[-1]
    
    def location(self, num):
        return self.deque.index(num)
        
N, M = map(int, input().split())
A = deque([ _ for _ in range(1, N+1)])
arr = list(map(int, input().split()))
count = 0

for i in arr:
    x = A.location(i)
    if x <= (A.len - x):
        for j in range(x):
            A.push_back(A.pop_front())
        count += x
        A.pop_front()
    else:
        for j in range(A.len - x):
            A.push_front(A.pop_back())
        count += A.len - x
        A.pop_front()

print(count)
```
