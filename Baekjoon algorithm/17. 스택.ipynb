```python
# 4949
class stack:
    def __init__(self):
        self.len = 0
        self.list = []
        
    def push(self, num):
        self.list.append(num)
        self.len += 1
    
    def pop(self):
        if self.len == 0: return -1
        self.len -= 1
        return self.list.pop()
    
    def size(self):
        return self.len
    
    def top(self):
        if self.len == 0: return -1
        else: return self.list[-1]
    
    def empty(self):
        if self.len == 0: return 1
        else: return 0

while True:
    A = stack()
    arr = list(input())
    if len(arr) == 1 and arr[0] == '.': break
    
    for i in range(len(arr)):
        if arr[i] in '([': A.push(arr[i])
        elif arr[i] in ')]':
            if arr[i] == ')': 
                if A.pop() != '(':
                    print('no')
                    break
            else:
                if A.pop() != '[':
                    print('no')
                    break
    
    if A.len == 0 and i == len(arr)-1: print('yes')
    elif A.len != 0 and i == len(arr)-1: print('no')
```

```python
# 1874
class stack:
    def __init__(self):
        self.len = 0
        self.list = []
        
    def push(self, num):
        self.list.append(num)
        self.len += 1
    
    def pop(self):
        if self.len == 0: return -1
        self.len -= 1
        return self.list.pop()
    
    def size(self):
        return self.len
    
    def top(self):
        if self.len == 0: return -1
        else: return self.list[-1]
    
    def empty(self):
        if self.len == 0: return 1
        else: return 0

N = int(input())
arr = []
solv = []
maxi = 0
A = stack()

for i in range(N):
    arr.append(int(input()))
    
for i in range(N):
    if arr[i] <= A.top():
        if arr[i] == A.top(): 
            A.pop()
            solv.append('-')
        else: 
            print('NO')
            break
    
    else:
        for j in range(maxi+1, arr[i]+1):
            A.push(j)
            solv.append('+')
        A.pop()
        solv.append('-')
    maxi = max(maxi, arr[i])
    
if i == (N-1):
    for i in solv: print(i)
```

```python
# 17298
class stack:
    def __init__(self):
        self.len = 0
        self.list = []
        
    def push(self, num):
        self.list.append(num)
        self.len += 1
    
    def pop(self):
        if self.len == 0: return -1
        self.len -= 1
        return self.list.pop()
    
    def size(self):
        return self.len
    
    def top(self):
        if self.len == 0: return -1
        else: return self.list[-1]
    
    def empty(self):
        if self.len == 0: return 1
        else: return 0

N = int(input()) 

arr = list(map(int, input().split())) 
A = stack()
solu = [-1 for _ in range(N)]

A.push(0)
i = 1

while A.len and i < N: 
    while A.len and arr[A.top()] < arr[i]:
        solu[A.top()] = arr[i]
        A.pop()
        
    A.push(i) 
    i += 1
        
for i in range(N): 
    print(solu[i], end = " ")
```
